```python
#Find a string

def count_substring(string, sub_string):
    k = len(sub_string)
    occ = 0                                              #occurencies of sub_string in string
    i = 0

    while i < len(string) - k +1:
        if string[i:i+k] == sub_string:
            occ += 1
        i+=1

    return occ
```

```python
#Text Wrap

def wrap(string, max_width):
    new = ''
    for i in range(len(string) // max_width) :
        new += string[i*max_width : (i+1)*max_width] + '\n'
    
    
    if not len(string) % max_width == 0:
        new += string[- (len(string) % max_width) :]
        
    return new
```

```python
#Alphabet Rangoli

def print_rangoli(size):

    import string

    length = (size-1) * 4 + 1

    #upper half
    for i in range(1, size):
        fake_line =  string.ascii_lowercase[ size - 1: size - i : -1 ] + string.ascii_lowercase[ size - i : size ]
        line = '-'.join([x for x in fake_line])
        print(line.center(length, '-'))
    
    #central line
    fake_line = string.ascii_lowercase[ size -1 :  : -1 ] + string.ascii_lowercase[ 1 : size ]
    print('-'.join([x for x in fake_line]).center(length, '-'))


    #lower half
    for i in range(size-1 , 0, -1):
        fake_line =  string.ascii_lowercase[ size - 1: size - i : -1 ] + string.ascii_lowercase[ size - i : size ]
        line = '-'.join([x for x in fake_line])
        print(line.center(length, '-'))

    return
```

```python
#The Minion Game

def minion_game(string):

    vowels = 'AEIOU'
    consonants = 'BCDFGHJKLMNPQRSTVWXYZ'

    n = len(string)
    c_words = 0
    v_words = 0

    for i in range(n):
        if string[i] in consonants:
            c_words += n-i   #n-i is the number of all possible substrings starting from s[i]
        
        if string[i] in vowels:
            v_words += n-i

    if c_words > v_words :
        print('Stuart', c_words )
    elif c_words < v_words :
        print('Kevin', v_words )
    else: 
        print('Draw')
    
    return
```

```python
#Merge the Tools!

def merge_the_tools(string, k):
    m = len(string)//k
    for i in range(m):
        t = string[i*k : (i+1)*k] #sliicing sbagliato
        print(t)
        print(delete_duplicates(t))
    
    return

def delete_duplicates(string):
    no_duplicates = ''
    for e in string:
        if e not in no_duplicates:
            no_duplicates += e
    
    return no_duplicates
```

```python
#collections.deque()

from collections import deque

d = deque()
N = int(input())     # lines of code following in input

for _ in range(N):
    command_line = input().split()
    if 'pop' in command_line:
        d.pop()
    
    if 'popleft' in command_line:
        d.popleft()
        
    if 'clear' in command_line:
        d.clear()
    
    if 'append' in command_line:
        d.append(command_line[1])

    if 'appendleft' in command_line:
        d.appendleft(command_line[1])
    
    if 'count' in command_line:
        d.count(command_line[1])
    
    if 'extend' in command_line:
        d.extend(command_line[1])
    
    if 'extendleft' in command_line:
        d.extendleft(command_line[1])
    
    if 'remove' in command_line:
        d.remove(command_line[1])

    if 'reverse' in command_line:
        d.reverse()
    
    if 'rotate' in command_line:
        d.rotate(n = command_line[1])
    
    print(' '.join([e for e in d]))
```

```python
#Piling Up!

from collections import deque

T = int(input())                            #number of test cases

for _ in range(T):
    n = int(input())                         #number of cubes
    blocks = deque(list(map(int, input().split())))
    
    pile = 'Yes'
    prev = 2**31     #cubes' length <=2**31 (hackerrank bound)
    while len(blocks) > 0:
        if blocks[0] >= blocks[-1] and prev >= blocks[0]:
            prev = blocks[0]
            blocks.popleft()
        
        elif blocks[-1] >= blocks[0] and prev >= blocks[-1]: 
            prev = blocks[-1]
            blocks.pop()
        
        else:
            pile = 'No'
            break

    print(pile)

```

```python
#Detect Floating Point Number
import re


def refloat(f):
    if re.search(r'.*[^0-9+\-\.]', f):    #if f contains any characters except 
                                          #the ones that create a float 
                                          #(<=> if float() raises an error
        return False
    elif re.search(r'.*[\+-].*', f[1:]):   #+- can only be in 1st position (indexed by 0)
        return False
    elif not re.search(r'.*\.+[0-9]', f):  #at least one decimal value
        return False
    elif len(re.findall(r'\.', f ))!=1: #exactly one point
        return False
    else:
        return True


T = int(input())      #number of test cases

for _ in range(T):
    f = input()
    print(refloat(f))
```

```python
#HTML Parser - Part 1
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Start :", tag)
        for attribute in attrs:
            print('->', attribute[0], '>', attribute[1])
    def handle_endtag(self, tag):
        print("End   :", tag)
    def handle_startendtag(self, tag, attrs):
        print("Empty :", tag)
        for attribute in attrs:
            print('->', attribute[0], '>', attribute[1])
    def handle_comment(self, data):
        pass


parser = MyHTMLParser()


N = int(input())   #lines of code

for _ in range(N):
    parser.feed(input())

parser.close()
```

```python
#HTML Parser - Part 2
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if '\n' in data:
            print('>>> Multi-line Comment', '\n'+data)
        else:
            print('>>> Single-line Comment', '\n'+data)
    
    def handle_data(self, data):
        if data != '\n':
            print( '>>> Data                 ', '\n'+data)
        
        
html = ""
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#Validating UID

import re

def repetition(l): 
    n = len(l)

    if n > 1:
        i = 1
        while i < n:
            if l[i] in l[:i]:
                return True
            i+=1
    
    return False

N = int(input())   #number of test cases

for _ in range(N):
    s = input()

    digits = re.findall(r'[0-9]{1}', s)
    upper = re.findall(r'[A-Z]{1}', s)
    lower = re.findall(r'[a-z]{1}', s)

    notalphanum = re.search( r'.*[^0-9a-zA-Z]', s)

    if len(s) != 10:
        print('Invalid')
    elif len(digits) < 3 or repetition(digits):
        print('Invalid')
    elif len(upper) < 2 or repetition(upper) or repetition(lower):
        print('Invalid')
    elif notalphanum:
        print(notalphanum)
        print('Invalid')
    else:
        print('Valid')
```

```python
#Validating Credit Card Numbers 

import re

pattern1 = r'(^[456])([0-9]{15}$)'
pattern2 = r'(^[456])([0-9]{3}\-[0-9]{4}\-[0-9]{4}\-[0-9]{4}$)'
pattern_rep = r'.*(1111|2222|3333|4444|5555|6666|7777|8888|9999|0000).*' 

N = int(input())    #number of test cases

for _ in range(N):
    card_number = input()

    if (re.match(pattern1, card_number) or re.match(pattern2, card_number)) and not re.match(pattern_rep, card_number.replace("-","") ):
            print('Valid')
    else: 
        print('Invalid')
```
